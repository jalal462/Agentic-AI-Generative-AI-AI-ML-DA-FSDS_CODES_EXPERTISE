```python
i=1
while i<=4:
    print('Data Science',end=" ") #when we mention end new line will not create
    j=1
    while j<=3:
        print('Technology', end=" ")
        j=j+1
    i=i+1
    print()
    
```

```python
i = 1
while i <= 4 :
    j = 0
    while  j <= 3 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```

# while else loop

```python
i = 10

while i <= 4 :
    j = 1
    while  j <= 3 :
        print(i*j, end=" ")
        j += 1
         
    print()
    i += 1
    
else:
    print('condition is not match')
```

# Break,Continue,Pass 

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(2,15):
    print(i)
```

```python
for i in range(1,11): # It will break the loop
    if i==4:
        break
    print(i)
```

```python
for i in range(2,21):
    if i==11:
        break
    print(i)
```

```python
for i in range(1,11): #skips the positions
    if i==5:
        continue
    print(i)
```

```python
for i in range(0,9):
    if i==6:
        continue
    print(i)
```

```python
for i in range(1,11): #skip the error
    pass
```

```python
x=int(input('How many chocolate do you want?'))

i=1
while i<=x:
    print('chocolate:',i)
    i+=1
    
```

```python
available_chocolate=8 #the machine only has 8 chocolate
x=int(input('How many chocolate do you want:'))

i=1
while i<=x:
    if i>available_chocolate:
        print('out of stock')
        break
    print('chocolate',i)
    i+=1
print('bye for now')
```
