# univariant analysis
##### plot graph using 1 variable

```python
vis1 = sns.distplot(stats["InternetUsers"])
plt.show()
```

```python
vis2 = sns.distplot(stats["InternetUsers"],bins = 10)
plt.show()
```

# bivariant analysis
##### plot the graph using 2 variables

```python
vis3 = sns.boxplot(data = stats,x= "IncomeGroup",y='BirthRate')
plt.show()
```

```python
vis = sns.lmplot(data = stats,x= "InternetUsers",y='BirthRate')
plt.show()
```

```python
vis = sns.lmplot(data = stats,x= "InternetUsers",y='BirthRate',fit_reg =False)
plt.show()
```

```python
vis = sns.lmplot(data = stats,x= "InternetUsers",y='BirthRate',
                 fit_reg =True,hue ='IncomeGroup')
plt.show()
```
